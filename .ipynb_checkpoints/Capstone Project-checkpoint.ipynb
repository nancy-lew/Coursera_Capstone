# Week 3 - Segmenting and Clustering Neighborhoods in Toronto

```python
#Scrapte data from website 

    # Webpage url                                                                                                               
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'

    # Extract tables
dfs = pd.read_html(url)

    # Get first table                                                                                                           
df = dfs[0]

    # Extract columns                                                                                                           
T = df[['Postal Code','Community','Neighbourhood']]
T
```

```python

#drop Not assigned in community (= borough)

T = T[~T['Community'].isin(['Not assigned'])]
T
```

```python
#if no same community, postal code = merge neighborhood
T = T.groupby(['Postal Code','Community'])['Neighbourhood'].apply(','.join).reset_index()
```

```python
T.shape
```

```python
#Import latlong
latlong = pd.read_csv("Geospatial_Coordinates.csv")
```

```python
latlong
```

```python
#join by postal code as index
newT = T.set_index('Postal Code').join(latlong.set_index('Postal Code'))
newT
```
